## Convolutional Neural Networks

### Your first CNN - __init__ method

<div class=""><p>You are going to build your first convolutional neural network. You're going to use the <code>MNIST</code> dataset as the dataset, which is made of handwritten digits from 0 to 9. The convolutional neural network is going to have 2 convolutional layers, each followed by a <code>ReLU</code> nonlinearity, and a fully connected layer. We have already imported <code>torch</code> and <code>torch.nn</code> as <code>nn</code>. Remember that each pooling layer halves both the height and the width of the image, so by using <code>2</code> pooling layers, the height and width are <code>1/4</code> of the original sizes. <code>MNIST</code> images have shape <code>(1, 28, 28)</code></p>
<p>For the moment, you are going to implement the <code>__init__</code> method of the net. In the next exercise, you will implement the <code>.forward()</code> method.</p>
<p><em>NB: We need <strong>2</strong> pooling layers, but we only need to <strong>instantiate a pooling layer once</strong>, because each pooling layer will have the same configuration. Instead, we will use <code>self.pool</code> twice in the next exercise.</em></p></div>

Instructions
<ul>
<li>Instantiate two convolutional filters: the first one should have <code>5</code> channels, while the second one should have <code>10</code> channels. The <code>kernel_size</code> for both of them should be <code>3</code>, and both should use <code>padding=1</code>. Use the names of the arguments (instead of using <code>1</code>, use <code>padding=1</code>).</li>
<li>Instantiate a <code>ReLU()</code> nonlinearity.</li>
<li>Instantiate a max pooling layer which halves the size of the image in both directions.</li>
<li>Instantiate a fully connected layer which connects the units with the number of classes (we are using <code>MNIST</code>, so there are <code>10</code> classes).</li>
</ul>

```
class Net(nn.Module):
    def __init__(self):
        super(Net, self).__init__()
        
        # Instantiate two convolutional layers
        self.conv1 = nn.Conv2d(in_channels=1, out_channels=5, kernel_size=3, padding=1)
        self.conv2 = nn.Conv2d(in_channels=5, out_channels=10, kernel_size=3, padding=1)
        
        # Instantiate the ReLU nonlinearity
        self.relu = nn.ReLU()
        
        # Instantiate a max pooling layer
        self.pool = nn.MaxPool2d(2, 2) 
        
        # Instantiate a fully connected layer
        self.fc = nn.Linear(7 * 7 * 10, 10)
```

### Your first CNN - forward() method

<div class=""><p>Now that you have declared all the parameters of your CNN, all you need to do is to implement the net's <code>forward()</code> method, and voila, you have your very first PyTorch CNN.</p>
<p>Note: for evaluation purposes, the entire code of the class needs to be in the script. We are using the <code>__init__</code> method as you have coded it on the previous exercise, while you are going to code the <code>.forward()</code> method here.</p></div>

Instructions
<ul>
<li>Apply the first <code>convolutional</code> layer, followed by the <code>relu</code> nonlinearity, then in the next line apply max-pooling layer.</li>
<li>Apply the second <code>convolutional</code> layer, followed by the <code>relu</code> nonlinearity, then in the next line apply max-pooling layer.</li>
<li>Transform the feature map from <code>4</code> dimensional to <code>2</code> dimensional space. The first dimension contains the batch size <code>(-1)</code>, deduct the second dimension, by multiplying the values for <code>height</code>, <code>width</code> and <code>depth</code>.</li>
<li>Apply the <code>fully-connected</code> layer and return the result.</li>
</ul>

```
def forward(self, x):

    # Apply conv followd by relu, then in next line pool
    x = self.relu(self.conv1(x))
    x = self.pool(x)

    # Apply conv followd by relu, then in next line pool
    x = self.relu(self.conv2(x))
    x = self.pool(x)

    # Prepare the image for the fully connected layer
    x = x.view(-1, 7 * 7 * 10)

    # Apply the fully connected layer and return the result
    return self.fc(x)
```

## Training Convolutional Neural Networks

## Training CNNs

<p>Similarly to what you did in Chapter 2, you are going to train a neural network. This time however, you will train the CNN you built in the previous lesson, instead of a fully connected network. The packages you need have been imported for you and the network (called <code>net</code>) instantiated. The cross-entropy loss function (called <code>criterion</code>) and the Adam optimizer (called <code>optimizer</code>) are also available. We have subsampled the training set so that the training goes faster, and you are going to use a single epoch.</p>

```
import torch.optim as optim
net = Net()
optimizer = optim.Adam(net.parameters(), lr=3e-4)
criterion = nn.CrossEntropyLoss()
```

```
net.forward = forward.__get__(net)
```

```
import torchvision
import torchvision.transforms as transforms

from six.moves import urllib
opener = urllib.request.build_opener()
opener.addheaders = [('User-agent', 'Mozilla/5.0')]
urllib.request.install_opener(opener)

# Transform the data to torch tensors and normalize it 
transform = transforms.Compose([transforms.ToTensor(),
								transforms.Normalize((0.1307), ((0.3081)))])

# Prepare the datasets
trainset = torchvision.datasets.MNIST('mnist', train=True, 
									  download=True, transform=transform)
testset = torchvision.datasets.MNIST('mnist', train=False, 
									  download=True, transform=transform)

# Prepare the dataloaders
train_loader = torch.utils.data.DataLoader(trainset, batch_size=1,
										  shuffle=True, num_workers=0)
test_loader = torch.utils.data.DataLoader(testset, batch_size=1,
										 shuffle=False, num_workers=0)       
```

Instructions
<ul>
<li>Compute the predictions from the <code>net</code>.</li>
<li>Using the <code>predictions</code> and the <code>labels</code>, compute the loss function.</li>
<li>Compute the gradients for each weight.</li>
<li>Update the weights using the <code>optimizer</code>.</li>
</ul>

```
for i, data in enumerate(train_loader, 0):
    inputs, labels = data
    optimizer.zero_grad()

    # Compute the forward pass
    outputs = net(inputs)
        
    # Compute the loss function
    loss = criterion(outputs, labels)
        
    # Compute the gradients
    loss.backward()
    
    # Update the weights
    optimizer.step()
```

### Using CNNs to make predictions

<div class=""><p>Building and training neural networks is a very exciting job (trust me, I do it every day)! However, the main utility of neural networks is to make predictions. This is the entire reason why the field of deep learning has bloomed in the last few years, as neural networks predictions are extremely accurate. On this exercise, we are going to use the convolutional neural network you already trained in order to make predictions on the <code>MNIST</code> dataset.</p>
<p>Remember that <code>torch.max()</code> takes two arguments:
-<code>output.data</code> - the tensor which contains the data.</p>
<ul>
<li>Either <code>1</code> to do <code>argmax</code> or <code>0</code> to do <code>max</code>.</li>
</ul></div>

Instructions
<ul>
<li>Iterate over the given <code>test_loader</code>, saving the results of each iteration in <code>data</code>.</li>
<li>Get the image and label from the <code>data</code> tuple, storing the results in <code>image</code> and <code>label</code>.</li>
<li>Make a forward pass in the net using your <code>image</code>.</li>
<li>Get the net prediction using <code>torch.max()</code> function.</li>
</ul>

```
net.eval()
# Iterate over the data in the test_loader
for i, data in enumerate(test_loader):
    # Get the image and label from data
    image, label = data
    
    # Make a forward pass in the net with your image
    output = net(image)
    
    # Argmax the results of the net
    _, predicted = torch.max(output.data, 1)
    
    if predicted == label: #all(predicted == label)
        print("Yipes, your net made the right prediction " + str(predicted))
    else:
        print("Your net prediction was " + str(predicted) + "\n    but the correct label is: " + str(label))
```
